```python
# Bibliotecas necessárias

from random import sample
import pandas as pd
```

```python
class MegaDaVirada():

        def __init__(self, dezenas, totalJogos):

            if dezenas >=6 or dezenas <=10: # condição
                self.__dezenas = dezenas
            else:
                print('Ops:/, insira dezena de 6 a 10')
        
            self.__totalJogos = totalJogos   # atributos privados 
            self.__resultado = None 
            self.__jogos = None


        @property
        def dezenas(self):
            return self.__dezenas

        @dezenas.setter
        def dezenas(self, value):
            if dezenas >= 6 or dezenas <=10:
                self.__dezenas = value
            else:
                print('Digite um valor entre 6 a 10.')

        @property
        def totalJogos(self):
            return self.__totalJogos

        @totalJogos.setter
        def totalJogos(self, value):
            self.__totalJogos = value

        @property
        def resultado(self):
            return self.__resultado

        @resultado.setter
        def resultado(self, value):
            self.__resultado = value

        @property
        def jogos(self):
            return self.__jogos

        @jogos.setter
        def jogos(self, value):
            self.__jogos = value
            

        def __Array(self):
            num = self.__dezenas
            num_sorteados = sample(range(1,61), num)
            num_sorteados.sort()
            return num_sorteados

        def sorteio(self):
            return self.__Array()

        def realizarApostas(self):
            total_jogos = self.__totalJogos
            jogo = []
            for i in range(0,total_jogos):
                jogo.append(self.sorteio())
                self.__jogos = jogo
                

        def realizarSorteio(self):
            sorteio = sample(range(1,61), 6)
            sorteio.sort()
            self.__resultado = sorteio

        # criando um data frame para retornar os valores 
        
        def conferirNumeros(self):
            nossosNumeros = self.__jogos
            dados = pd.DataFrame()
            resultado = self.__resultado   
            num_jogo = []
            numeros = []
            resultado_sorteo = []
            qnt_acertos = []
            num_sorteados = []

            for i in range(0, len(nossosNumeros)): 
                num_jogo.append(i+1)
                numeros.append(str(nossosNumeros[i]).strip('[]'))
                resultado_sorteo.append(str(resultado).strip('[]'))
                qnt_acertos.append(len(set(nossosNumeros[i]).intersection(resultado)))
                num_sorteados.append(0 if str(list(set(nossosNumeros[i]).intersection(resultado))).strip('[]') == '' else \
                                    str(list(set(nossosNumeros[i]).intersection(resultado))).strip('[]'))
                    
            dados['Jogo Nº'] = num_jogo
            dados['Meus números'] = numeros
            dados['Dezenas Sorteadas'] = resultado_sorteo
            dados['Quantidade de acertos'] = qnt_acertos
            dados['Meus números sorteados'] = num_sorteados
            return dados



dezenas = int(input('Digite a quantidade de dezenas: '))

while dezenas < 6 or dezenas > 10:
    print('Digite um valor entre 6 a 10.')
    dezenas = int(input('Digite a quantidade de dezenas: '))

else:
    pass

        
totalJogos = int(input('Digite a quantidade total de jogos: '))
print('\n')


MegaDaVirada = MegaDaVirada(dezenas, totalJogos) # criando o objeto


escolha = input('Escolha uma das opções:\n1 = Realizar Aposta\n2 = Realizar Sorteio\n3 = Conferir Resultados\n4 = Sair\n')


while escolha != '4':

    if escolha == '1':
        MegaDaVirada.realizarApostas()
        print('Seus números:')
        print(MegaDaVirada.jogos)
        print('\n')

        escolha = input('Escolha uma das opções:\n1 = Realizar Aposta\n2 = Realizar Sorteio\n3 = Conferir Resultados\n4 = Sair\n')

    elif escolha =='2':
        MegaDaVirada.realizarSorteio()
        print('Números sorteados:')
        print(MegaDaVirada.resultado)
        print('\n')

        escolha = input('Escolha uma das opções:\n1 = Realizar Aposta\n2 = Realizar Sorteio\n3 = Conferir Resultados\n4 = Sair\n')

    elif escolha == '3':
        print('Resultados:')
        print(MegaDaVirada.conferirNumeros())
        print('\n')
            
        
        print('Obrigado!')
        break
    elif escolha == '4':
        print('Obrigado!')
        break
        
```
